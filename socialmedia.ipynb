```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("twitter_training.csv")
df.head()
```

```python
header=['ID','Entity','Sentiment','Message']
df.columns=header
df.head()
```

```python
df.to_csv('twitter.csv',index=False)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Message'])
df.isnull().sum()
```

```python
df['Sentiment'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Sentiment',palette='Set2')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
sentiment_entity_counts = df.groupby(['Sentiment', 'Entity']).size().reset_index(name='Count')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=sentiment_entity_counts, x='Sentiment', y='Count', hue='Entity', palette='Accent')
plt.title('Sentiment Distribution by Entity')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
